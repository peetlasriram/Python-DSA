# String Algorithms -1
1. What is prefix, suffix, z-function, KMP
* These concepts helps use efficiently compare or search patterns in strings instead of checking character by character.

* Prefix
* Defination:- A prefix is the beginning part of string.
* Example:
* For "Abcde" Prefix are "a","ab","abc","abcd","abcde"

* suffix:
* Defination :- A suffi is ending of the part of string.
* For "abcde", Suffix are "e","de","cde","bcde","abcde"

* Z-array(Z-function)
* Defination:- For Each index i in a string, Z[i]= ;length of longest substring starting at i that matches prefix of the string.

* Example :
* For string S= "ababcab" , z[0]= 0 (not defined)
* But from index 2: S[2:] = "abcab" and it matches prefix "ab" --> Z[2]=2
* S[i:] ---> matches --> Z[i]
* ababcab --->[-] ----> 0
* babcab ---> no match---> 0
* abcab --> "ab"--> 2
* bcab -->"b"--> 0
* cab --> no match--> 0
* ab --> "ab--> 2
* b--> "b"---> 0

* Z-array is used in Z-Algorithm, which does pattern searching in O (n) time.

* Failure function pai-array (KMP Prefix Table)
* use in :- Knuth-morris-pratt (KMP) pattern matching.
* pai[i]-tells us: length of the laongest proper prefix of the pattens p[0...1] which is also a suffix, helps us avoid unnecessary -comparisons during pattern search.

* Example: For pattren="ababc"
* pai=[0,0,1,2,0]
* why 
* pai[0]=0(only one character)
* pai[1]=0(no prefix= suffix)
* pai[2]=1(a matches a)
* pai[3]=2(ab matches ab)
* pai[4]=0(no matching prefix and suffix)

2. When & why use efficient string matching algorithms
* why not use a simple loop
* Brute-force string matching is O(n*m)--> For each position, try to match the full pattern.

3. Efficient Algorithms
* Algorithms--> Uses ---> Time
* KMP --> Uses Pai-array---> O( n+m)
* Z-algorithms--> Uses Z-array-->O(n+m)
* Rabin-Karp--> hash-based --> O(n+m)
* Aho-Corasick --> Multi-pattern --> O(n+total pattern length)

4. Time and space complexity
* Algorithms--> Time --> Space---> Notes
* KMP --> O(n+m)---> O(m)---> Builds Pai-array
* Z-algorithm-->O(n+m)--->O(n+m)---> builds Z-array
* Rabin -Karp ---> O(n+m)--> O(1)---> May degrade in worst case
* Brute force---> O(n x m)---> O(1)---> Slowest

5. Advantages 
* Fast patterens in liner time.
* useful in competitive programmming , text processing
* Eliminates re-checking repeacted characters
* scalable for big data. logs, etc

6. Disadvantages
* Slightly complex to implement for biginners.
* Requires preprocessing
* Uses extra space (array like Z[],Pai[])

7. Final Summary
* Concepts ----> Used in ---> key ideas--> complexity
* Prefix--> Everywhere ---> Starting part of a string ---> -
* Suffix-->Everywhere ---> Ending part of a string--->-
* Z-aaray---> Z-Algorithm---> How much suffix matches the prefix --> O(n)
* pai-Array--> KMP---> prefix = Suffix mapping in pattern -->O(n)
