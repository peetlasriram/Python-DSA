# BellMan-Ford Algorithms in DSA
1. What is Bellman-Ford 
* The Bellman-ford algorithm is graph algorithm used to find the shortest paths from a single source node to all other nodes in a weighted graph, even if the graph contains negative edge weights.

2. How it works.
* Initialize distance to all nodes as (Expect the source =0)
* Relax all edges n-1 time (where n= number of vertices)
* After n-1 relaxation, all shortest paths will be found.
* optionally, run one or more pass to ditect a negatiive weights cycle

3. When to use Bellman-Ford
* Situation ---> Use Bellman ford if
* Graph has negative  edge weights --> Yes
* you need to detect a negative weight cycle --> yes
* you want shortest paths from one source to all nodes ---> yes
* Graph is small or medium- sized --> yes
* Graph is very large with only positive weights --> use Dijkstra instead (faster)

4. key terminologies 
* Term ---> Meaning
* Edges Relaxation --> the process of upadating the shortest distance to a node.
* Negative weight edge --> An edge with a negative cost ( e.g ,-5)
* Negative weight cycyle ---> A cycle where the total weight becomes negative forever
* Source node ----> The starting point from which all shortest paths are calculated.

5. time and space complexity 
* Time --> O(V * E) where (v = number of vertices),(E= number of edges)

* Why (We loop though all edges for v-1 times)

* Space complexity -->O(v) only one distance[] array to store shotest distances.

6. Advantages
* Feature ---> Explanation 
* Works withh negative weighs ---> Unlike Dijkstras
* Can detect negative weight cycle --> important in graph where weight can decrease over time.
* simle logic using basic loops --> No fancy data structure

7. Disadvantages
* Slower that Dijkstra ---> time complexity is higher, specially for dense graphs
* Not suitable for very large graphs ---> O(v+E) to bee Slow
* Doesn't work with undirected negative edges --> Because it leads to negative cycle  easily.

* Real-world uses Cases
* Currency Arbitrage ( Detecting negative cycle in exchange rates)
* Routing Algorithms where weights can be negative (e.g.with refunds or penalities)
* Network protocols where conditions may change dynamically.

8. Final Summary
* Aspect --> Bellman ford 
* Purpose ---> Shortest paths from a single Source 
* handle Negative --> yes
* Detect Cycles --> yes ( negative weight cycle)
* Time complexity --> O(VxE)
* space complexity ---> O(V)
* Better than Dijkstra only if negative or cycles are involved.





```python
# let's implement Bellman-ford algorithm in Python
def bellman_ford(n,edges, start):
    # Initialize distances from start to all other vertices as INFINITE
    distances = [float('inf')] * n
    distances[start] = 0

    # Relax all edges |V| - 1 times.
    for _ in range(n - 1):
        for u, v, w in edges:
            if distances[u] != float('inf') and distances[u] + w < distances[v]:
                distances[v] = distances[u] + w

    # Check for negative-weight cycles.
    for u, v, w in edges:
        if distances[u] != float('inf') and distances[u] + w < distances[v]:
            raise ValueError("Graph contains a negative-weight cycle")

    return distances
```

```python
# Example usage:
if __name__ == "__main__":
  n = 5  # Number of vertices
  edges = [
        (0, 1, -1),
        (0, 2, 4),
        (1, 2, 3),
        (1, 3, 2),
        (1, 4, 2),
        (3, 2, 5),
        (3, 1, 1),
        (4, 3, -3)
    ]
start = 0  # Starting vertex
distances = bellman_ford(n, edges, start)
print("Vertex Distance from Source")
for i in range(n):
      print(f"{i}\t\t{distances[i]}")
  
    
```
