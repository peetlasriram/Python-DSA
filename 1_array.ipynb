# What is an Array
An Array is Data structures that store collection of elements in the datatype in contignous block of memory, access via indices

# Advantages of Arrary 
1. Constact Time Access (O(1)): Fast Element Retrive
2. Memory efficient : Stored Contiguosly with Low over Head.
3. Efficient Traversal : Easy to iterate though element

# Common operations:
1. Traverasl: iterating though elements.
2. Access : Retrieving elements via Index.
3. Modification : Changing values at specific indices
4.  inserction/Delection : Adding/Removing elements (can be O(n) for mid array Operation).
5. Search : Finding the elements O(n) for unsorted array

# Time complexity

1. Access : O(1)
2. Search : o(n)
3. Inserction/Delection (end): O(1)
4. Inserction/Delection (middle): O(n)
5. sorting :O (n log n)

1. Find _Largest array

```python
# brute Forse Way 
def find_mx(arry):
    max_ele=arry[0]
    for num in arry:
        if num > max_ele:
            max_ele=num
    return max_ele
```

# Find Second Largest Element in Array

```python

# Brute_force timr Complexity O(n log n)
def second_loargest(array):
    soreted=sorted(array,reverse=True)
    return soreted[1]
    
    
```

```python
# optimal way O(n)
def second_largest_optimal(array):
    first=second=float('-inf')
    for num in array:
        if (num > first):
            second=first
            first=num
        elif num > second and num != first:
            second=num
    
    if second !=  float("-inf"):
        return second
    return "not found"
```

```python
arry=[1,7]
print("Second Largest :",second_largest_optimal(arry))
```

```python
arry=[1,2,3,4,5,6,7,7,6,8,9]
print("Second Largest :",second_loargest(arry))
```

# 2.Check if an array is sorted

```python
# Brute Force  O(n)
def brute(arr):
    for i in range(len(arr)-1):
        if (arr[i]> arr[i+1]):
            return False
    return True
```

```python
#Optimal way
def optimal(arr):
    return all(arr[i]< arr[i+1] for i in range(len(arr)-1))
```

```python
array=[1,2,3,4,5,6,7,]
print("sorted:",brute(array))
```

```python
array=[1,2,3,4,5]
print("sorted:",optimal(array))
```

# 3. Reverse the elements of an Array

```python
def brute_r(arr):
    n=len(arr)
    for i in range(n//2):
        arr[i],arr[n-1-i]=arr[n-1-i],arr[i]
    return arr
        
```

```python
def opitmal_r(arr):
    return arr[::-1]
    
```

```python
array=[1,2,3,4,5]
print("reverse the Element:",brute_r(array))
```

```python
array=[1,2,3,4,5]
print("reverse the Element:",opitmal_r(array))
```

# 4. Remove Duplicates from an Array

```python
def brute_rr(arr):
    une=[]
    for num in arr:
        if num not in une:
            une.append(num)
    return une
    
```

```python
def opitmal_rr(arr):
    return list(set(arr))
    
```

```python
array=[1,2,3,4,5,5,4,2]
print("remove duplicate the Element:",brute_rr(array))
```

```python
array=[1,2,3,4,5,5,4,2]
print("remove duplicate  Element:",opitmal_rr(array))
```

# 5. Rotate an Array by K Position

```python
def brute_(arr,k):
    n= len(arr)
    rotations= k % n
    for _ in range(rotations):
        last_ele=arr.pop()
        arr.insert(0,last_ele)
    return arr
    
```

```python
def opitmal_(arr,k):
    k = k % len(arr)
    return arr[-k :]+ arr[:-k]
    
    
    
```

```python
array=[1,2,3,4,5]
k=2
print("rotate an array:",brute_(array,k))
```

```python
array=[1,2,3,4,5]
k=2
print("rotate an array:",opitmal_(array,k))
```

# 6. Move Zeros(Leet code Q283)

```python
def brute_Z(arr):
    n=len(arr)
    non_zeros=[num for num in arr if num != 0]
    zero=n-len(non_zeros)
    reslut=non_zeros+[0]*zero
    for i in range(n):
        arr[i]= reslut[i]
    return arr
```

```python
def opitmal_Z(arr):
    poniter=0
    for i in range(len(arr)):
        if (arr[i]!=0):
            arr[poniter],arr[i]=arr[i],arr[poniter]
            poniter+=1
    return arr     
```

```python
arr=[1,0,2,3,0,4]
print("move zeros End:",brute_Z(arr))
```

```python
arr=[1,0,2,3,0,4]
print("move zeros End:",opitmal_Z(arr))
```

# 7. Find the missing Number (Leetcode Q268)

```python
def brute_F(arr):
    for i in range(len(arr)+1): # O(n)
        if i not in arr:
            return i
```

```python
def opitmal_F(arr):
    
    n=len(arr)
    exp_sum = n*(n+1)//2
    acutal=sum(arr)
    
    return exp_sum - acutal
    
    
```

```python
arr=[0,2,3,0,4]
print("missing Number:",brute_F(arr))
```

```python
arr=[1,3,4,5,7,9]
print("missing number:",opitmal_F(arr))
```

# 8. Intersection of Two array (leet code Q349)

```python
def brute_I(arr,a):
    reslut=[]
    m=len(arr)
    n=len(a)
    if m > n:
        for num in arr:
            if num in a and num not in reslut:
                reslut.append(num)
    else:
        for num in a:
            if num in arr and num  not in reslut:
                reslut.append(num)
    return reslut
    
```

```python
def opitmal_I(arr,arr1):
    return list(set(arr)&set(arr1))
```

```python
arr=[0,2,3,0,4]
a=[1,4,3,5]
print("intersection number :",brute_I(arr,a))
```

```python
arr=[1,3,4,5,7,9]
a=[1,4,5]
print("missing number:",opitmal_I(arr,a))
```

# 9. Find the Majority elements in Array(leetcode Q169)

```python
def brute_FM(arr):
    for x in set(arr):
        if arr.count(x) > len(arr)//2:
            return x
    return "not found"
```

```python
def opitmal_FM(arr):
    count=0
    x =None
    for num in arr:
        if count == 0:
            x= num
        count+=1 if num == x else -1
    return x
```

```python
arr=[3,2,3]
print("Majority elements :",brute_FM(arr))
```

```python
arr=[2,2,1,1,1,1,4,5]
print("Majority Elements :",opitmal_FM(arr))
```
