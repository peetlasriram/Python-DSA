#   Basic Operation on strings
a. Concation
Combining two or more strings using + operator:


```python
Str1="Hello"
str2="World"
reslut=Str1 + " " + str2
print(reslut)
```

b. Slicing
Extracting Specific parts of string using slicing 


```python
s="python programming"
print(s[0:6])  # prints 'python'
print(s[-11:])   # prints 'programming
print(s[::-2])    # prints 'python'
```

c.immutability
Strings are immutable , Modifing a string create a new string 

```python
s="python"
s[1]='a'  # This will raise an error
```
