# 1. Reverse a String  leetcode Q:

```python

def bruteForceRs(s):
    reversed_s = ''
    for char in s:
        reversed_s = char + reversed_s
    return reversed_s
```

```python
def opitimalRs(s):
    return ''.join(reversed(s))
```

```python
def opitimalbs(s):
    return s[::-1]
```

```python
def reverse(s):
    n = len(s)
    for i in range(n// 2):
        s[i],s[n-i-1] = s[n-i-1],s[i]
    return s
```

```python
s="hello world"
print("Brute Force Sting:",bruteForceRs(s))
print("opitimized Sting:",opitimalRs(s))
print("opitimized Sting with slicing:",opitimalbs(s))
print("reverse:",reverse(s))
```

# 2. valid palindrome  leetcode Q:125

```python
def bruteforceVp(s):
    new_s=""
    rev_s =""
    for char in s:
      for i in char:
         if i.isalpha():
            new_s+= i.lower()

            
    for c in new_s:
      rev_s= c+ rev_s
    return new_s == rev_s
```

```python
def opitimalVp(s):
    new_s=" ".join(char.lower() for char in s if char.isalnum())
    lis=list(new_s)[::-1]
    revs="".join(lis)
    return new_s  == revs 
```

```python
s="madam"
print("brute Force method for Valid palindrome:",bruteforceVp(s))
print("opitimized method of valid palindrome:", opitimalVp(s))
```

# 3. count the number of vowel String in range leet code Q 2586

```python
def bruteCNVS(strings,left,right):
   result=0
   vowel={"a","e","i","o","u"}
   for i in range(left,right+1):
      strin=strings[i]
      if strin[0] in vowel and strin[-1] in vowel:
         result+=1
   return result
```

```python
def opitimalCNVS(words,left,right):
    vowel={"a","e","i","o","u"}
    return sum(1 for i in range(left,right)if words[i][0] in vowel and words[i][-1] in vowel)
```

```python
stings=["are","amy","u"]
left=0
right= 2

print("Brute Force techinue for number of vowel Strings in a range:",bruteCNVS(stings,left,right))
print("opitimal  techinue for number of vowel Strings in a range:", opitimalCNVS(stings,left,right))
```

# 4. valid angram (leetcode Q242)

```python
def bruteAG(m,n):
    return sorted(m)==sorted(n)
```

```python
def opitimalAG(m,n):
    cc={}
    if len(m) != len(n):
        return False
    for i in m:
        cc[i]=cc.get(i,0)+1
    print(cc)
    for j  in n:
        if  j not in cc or cc[j] == 0:
            return False
        cc[j]-=1
    print(cc)

    return all(count==0 for count in cc.values())

```

```python
m="angram"
n="ngaram"
print("brute force techinue for angram:",bruteAG(m,n))
print(" opitimal techinue for angram:",  opitimalAG(m,n))
```

# 10. First unique character in a string leetcode Q387

```python
def bruteFUC(s):
    for i in range(len(s)):
        if s.count(s[i])==1:
            return i
    return -1
```

```python
def opitimalFUC(s):
    cc={}
    for char in s:
        cc[char]=cc.get(char,0)+1
    
    for i, char in enumerate(s):
        if cc[char]==1:
            return i
    return -1

```

```python
s="leetcode"
print("brute Force techinue for first unique character in a string:",bruteFUC(s))
print("opitimal techinue for first unique character in a string:",opitimalFUC(s))
```
