# Hashing
1. What is Hashing 
a. Hashing is process of mapping data from a large (like string or number) to 
   a smaller. fixed-sized value called a hash code. This is done using a hash function, which take the input and convert it into an integers (hash code) that ideally represent the data. this integers is then used as an index for storing or retriving the data ina data structure, such as hash table.

   The key idea behind hashing is to allow for efficient lookup, inserction, and deletion by using this mapping.

   Difference between Hashmap And Hashset(python) 

2. Hashmap(Dictionary in python )
a. Definition: A hashmap (called a dictionary in python) is a collection of    key-values pairs, Each a key is associated with a specific value.

Use: You can retrive , update, or delete values based on a specific key
 
 Example:
 my_dict={"a":1,"b":2}

print(my_dict["a"])

operatons: insertion,deletion,searching

Use: you can check if an item is present in the set, add new items, or remove existing ones.

Example:

operation: inserction ,deletion, membering check.

usecases and advantages in data structure and algorithms (DSA)

Efficient Lookup: Hashing provides constant- time complexity(0{1}) for lookups, inserctions, and deleted

Fast Access: Hashing is ideal when you need fast access to data based on a unique identifier, such as dictionary lookup, caching. and data indexing.

Unique Elemennts:Hashset ensures that there are no duplicates, making them useful when you need a collection of unique items, such as checking for duplicates in a list,

Efficient searching: Hashing is used in algorithms like hash table or hash map for quick searches or lookup which can improve overall performance of algorithms.

cache implemention: hashing is widely used for implementing caching system, where you store compoted result to avoid redundant, enhancing speed.

Handling collision 

Collision occur when two different input(key) generate the same hash value (same index)

Commom method to handle collisions:
chaining: Store multiple element at the same index (usually in linked list or anothe structure).
open addressing: when a collisions occurs, probe (search) for the next available slot in the table.

ex: (open addressing approach):

if two keys hash to the same index, you might try the next index, and so on untill an empty spot is found.

# 1. Find first  non- repeating element in an array

```python
def bruteff(arr):
    for i in range(len(arr)):
        unique=True
        for j in range(len(arr)):
            if i != j and arr[i]==arr[j]:
                unique=False
                break
        if unique:
            return arr[i]
    return -1

arr=[9,4,9,6,7,4]
print(bruteff(arr))
```

```python
def optimalff(arr):
    fer={}
    for num  in arr:
        fer[num]=fer.get(num,0)+1
        print(fer)
    for num in arr:
        if fer[num]==1:
            return num
        
    return -1

arr=[9,4,9,6,7,4]
print(optimalff(arr))
```

```python
def bitlsn(arr):
    result=0
    for num in arr:
        result ^= num
    return result
arr=[1,1,2,4,2,3,3]
print(bitlsn(arr))
```
