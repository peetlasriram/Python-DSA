# Greedy Algorithms
1. What is Greedy Algorithm
* A Greedy Algorithm is a method where:
* at each step. You make the best choice(the most "greedy "choice) without worring about the future.

2. You only think
* "what looks best right now"
* Goal: By picking the best opition at every step. You hope to get the best overall solution.

3. Why is it called Greedy Algorithm
* Because at every decision point.
* you grap the best immediate opition- without thinking about thr bigger picture, Like a greedy person who grabs the biggest piece of cake without planning for other.

4. How greedy algorithms works
* Start from the first point.
* At each step, Pick the best possible choice.
* Move to the next step based on that choice.
* Repact until the end.
* you never go back to fix your previous choice.

5. Improtant Points:
* Greedy algorithms always don't give the prefect answer for every problem.
* But for some problems, Greddy gives the best (optimal) Solution.
* For other, It gives a good enough solution, but not perfect.

6. Real-Life Example (Easy to understand)
* Coin Chnage problems(Choosing largest coin first)
* Imegine you have to make 27 rupees  using of 25 10 5 and 1 rupees

7. Greedy Approach
* Pick the largest coin possible at each step.
* 25 (1 coin)-> 27 ->25== 2
* 1 (2 coins)->2-> 1->1== 0
* Total coins used : 1+2 = 3 coins
* Here ,greedy gives a good answer

8. When does the Greedy works prefectly
* Greedy work when:
* Greedy choice properity:
* choossing the best local opition also leads to the best global solutions.

9. Final Tips
* Always think making the best current choice will lead to the best final answer
* If yes- Greedy Works
* if no - You may need other methos( Like dynamic Programming)



```python
# let's implement Greedy algorithm to solve the Coin Change Problem
def greedy_coin_change(coins, amount):
    # Sort the coins in descending order
    coins.sort(reverse=True)
    
    result = []
    for coin in coins:
        while amount >= coin:
            amount -= coin
            result.append(coin)
    
    if amount != 0:
        return None  # No solution found
    
    return result
```

```python
# Example usage
if __name__ == "__main__":
    coins = [25, 10, 5, 1]
    amount = 27
    change = greedy_coin_change(coins, amount)
    if change is not None:
        print(f"Coins used to make change for {amount}: {change}")
    else:
        print("No solution found")
# to find the minimum number of coins needed to make a certain amount.  
```
