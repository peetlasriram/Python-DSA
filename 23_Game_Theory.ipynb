# Game Theory
1. What is a GameTheory
*  Game theory is the Study of decision -making between two or more playes.
* In DSA. Game theory helps us solve problems where:
* Two players take turns
* Each playe plays optimally (smartly)
* The goal is to determine the winner or maximum again
* you're not simulating the whole game--you're predicting the outcome using logic, recursion, and dynamic programming.

* Example Problem: Nim Game, Stone Game, Coin Game, Predict the winner.


2. When to use Game Theory
* Use Gametheory in DSA When:
* Two players play alternatively
* Each playes plays optimally (not ramdomly)
* the game's rules are fixed.

* you're asked to find 
* who will win
* what's the best move.
* what's the maximum/ minimum score possiable

* Key words in problems statements.
* "two players"
* "Take turns"
* "Play optimally"
* "Find who wins"

3. Key terminologies in gametheory
* Term     | Meaning
* Game State | the current situation (e.g. Remaining Coins, Stones)
* Move    | An action a player can take on their turn
* winning State | A game state where a player can force a win
* Lossing State | A game state where every move leads to awin for the opponent
* optimal play  | Always making the best possiable move
* minimax strategy | to minimize opponent's gain and maximize your own.

4. Time and Space complexity(General Idea)
* Most game theory problems are solved using
* 1. Recursion
* * Time : o(2*N)(exponential)
* * Space Complexity : o(n) (Stack Space )

* 2. Recursion+ Memorization (top-Down Dp)
* * Time o(n) or o(n*2) (based on state space)
* * Space complexity : O(n*2) (for memo table)

* 3. Bottom -up DP ( Tabulation)
* Time: O(n) or O(N^2)
* Space : O(n) or o(N^2)

* It depends on Size of Game state ( Array, Stones,turn.etc)
* Number of variable on each state (Position, turn,etc.)


5. Advantages 
* Predict outcome : you can predit if the first or second player will win.
* Optimize decisions : Helps in making smart moves instead of brute force.
* Learn DP : Most game theory problems are good partice for dynamic programming.
* Used in contests: Common competitions (Nim Game, Stone Game, etc)

6. Disadvantages
* Hard to identify : It's not always obvious that it's game theory problem
* Recursion can be slow : without memorization, it may lead to TLE (Time limit Exceeded)
* opitimal play is tricky : You need to think for both players
* State Explosion : Conplex game may have too many states
