# 1. What is Searching
* A: Searching is the process of finding an element in a given dataset. It is widely used ib programming to retrive data efficiently

# 2. Type of Searching Algorithms
 * Linear search
 * Checks each element one by one
 * Works well for small or unsorted data.
 * Time Complexity:
 * Best case: O(1) (Element found at the Start)
 * Worst Cae: O(N) (Element found at the end of the Present)

* Binary Search 
* Only works on sorted arrays.
* Uses a divide-and-Conquer approach by repeatly having the search space.
* Time Complexity:
* Best case: O(1) (middle elemenet is the target).
* Worst case: O(log N) (Continuously dividing search space) 

# 3. What is Sorting 
* A: Sorting is the process of arranging element in the specific order ( ascending or decending order). It is cruial in DSA Because many Advanced Algorithms (Like binary Search) Require Sorted data.

# 4. Why is sorting important
*  A: Enales fast searching (binary search) helps in solving problems like finding duplicates. Merging arrays etc. Optimizes algorithms like quickselect ( for finding the Kth Largest/Smallest elements=)

# Time Complexity Comparison od Sorting algorithms


# Bubble sort   
*   Best Case:O(N)  * Average Case:O(N2) * Worst Case: O(N2) 
# Selection Sort
* O(N2) * O(N2) * O(N2)
#  Insertion Sort
* O(n) * O(N2) * O(N2)
# Merge Sort
* O(N log N) * O(N log N) * O (N log N)
# Quick Sort
* O(N log N) * O (N log N) * O(N2)
# Heap Sort
* O (N log N) * O( N log N) * O(N log N)
# Counting Sort
* O(n+k) * O(n+k) * O(N+K)

# Key Take way
* Bubble sort, and insertion sort are easy but slow(O(n2)). Merge sort and quick sort are efficient (O(N log N)). Counting sort is the fastest but works only for small integers range.

# When to use which algoritm
* Use Liner Serach: when the dataset is small or unsorted. Searching once and sorting is not beneficial.

* Use Binary Search: When the dataset is sorted. you need multiple searches on the same datasets.

* Use Merge sort : When Stable sortting is required. You need a guaranteed O (N log N) Performance.

* Use Quick Sort: when the average case O(N log N), and extra is not preferable sorting happens in-Place(without additional memory overhead).

* Use Counting Sort: When the data range is limited (eg:Sorting integers between 1-1000). works best for integer -Based sorting.
