# Linked_List
1. What is a Linked_list.
* A linked list is a type of data structure used to store elements ina lineaar order. unlike an array, which store elements in continuous memory location, a linked list store elements at random locations and connect them using pointers. 
* Each element in a linked list is called a node, and each node contains: Data (The actual value), Pointer (or link ) to the  next node in the list.
2. types of linked_list (single, Doubly Circular)
3. Advantages & Disadvanages
4. Basic Nodes Structure & implemention

# Typs of linked list There are main type of linked list 
1. Singly linked list
* Each node has Data
* Pointer to the next node
* E g: A -> B-> C-> D-> None(Each node points to the next one , and the last node points to node)
2. Double Linked List
* Each node has Data
* Pointer to the next node
* Eg:<-A<-> B<-> c<->d-> None(Each node has links to both the previous and next nodes)
3. Circlular linked list
* In a circular single linked list: the last node points back to the first node.
* in a circular  double linked list :both last and first nodes are linked to the each other in the both directions.
* E g:(Circular singly linked list): A->B -> C->D-> A(back to A)


# Adavanages  & Disadvanates 
1. Feature           |   Linked list    | Array
* Memory Allocation  | Uses dynamic memory alllocation (can grow/shrink easily) | Fixed Size
* Insertion/deletion  |Fast(o(1)at beginning) | Slow(o(n)if shifting is needed)
* Access Time |Slow(o(n), must traverse from the head) | Fast (o(n),Direct index access)
* Extra Space | needs extra space for pointer |No extra space required
* Efficiency |Efficient for dynamic data | Efficiently for quick access

# When to use 
* Use linked list when you need frequently insertions/ Deletion then that time only use linked list
* use arrays when you need fast random access

```python
# Here's how to create a basic node and singly linked list in python
# Creating a node 

class Node:
    def __init__(self, data):
        self.data = data # data of the node
        self.next = None # pointer to the next node
# Creating a singly linked list
class SinglyLinkedList:
    def __init__(self):
        self.head = None # head of the list

    # Method to insert a new node at the end of the list
    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    # Method to display the linked list
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
```

```python
# Example usage
if __name__ == "__main__":
    sll = SinglyLinkedList()
    sll.insert(1)
    sll.insert(2)
    sll.insert(3)
    sll.display()  # Output: 1 -> 2 -> 3 -> None
```
