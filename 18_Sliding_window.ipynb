# --> Sliding Window <--
1. What is the sliding Window Technique
* Imagine you're looking though a window that shows only a few items at a time -- Like 3 eords in a sentence, 3 house in a row, or numbers in list.

* Instead of analyzing the entire list over and over. you focus on what's inside the window , and then you slide the window one step forward- to the next group of items.
* It's A way of breaking a big problem into a small parts, moving step by step, and reusing what you already know instead of starting from scratch each item.

* [1,2,3,4,5,6.7]

2. Real-life Analogy : Train window
* Picture yourself on a moving train, Looking out the window.
* you can only see a few houses at a time- let's say 3.
* As the train moves to forward, you don't go back to the beginning to look again.
* you just let go of the house that passed, and look at the new house that just appeared.
* That's how the sliding window works.
* you can process one group of items, slide forward, adjust what changed. and keep going - first and match.

3. When Do you use the sliding window
* you use it when the problems involve array or string thing in a row.
* you're working with a group of consecutive (next to each other) Elements.
* you're asked to find something like:
* The maximum/minimum (e.g .Longest/Shotest substring or subarray)
* whether a certain pattern or condition is meet
* instead of checking every possibility from the beginning (which is slow ),You just slide and update.


4. Two main types.
1. Fixed -size Window
* The size of window is set and doesn't change.
* Example: Find the highest Average of 3 test score ib a row
* you slide one element at a time . always looking at 3 times.
2. Variable -Size Window
* The window grows or shinks depending on conditions,
* Example: "Find the shotest part of a video that includes all improtant scenes.
* You expand untill you meet the condition, then shirks to make it as small as possiable.

5. Why is it better than brute force
* Brute force means checking every possibility- which is like walking to each house, writing down its' value and starting over each time. Its'slow.
* Sliding window means you remember what you just saw, And only update the part that changed- which makes it much faster in techniccal term. you can improve the spped from 0(N2) to 0(N) (Fast), which is a huge difference.
