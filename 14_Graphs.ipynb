# Graphs
1. What is Graph
* A graph is a way to show connections or relationship between things.
* Graph Has:
* Vertices( nodes): these are the "things" or "points"
* Edges ( Links) :These are the connections between the things 

2. Simple Example :
* imgine cities connected by roads:
* Cities=Vertices
* Roads== Edges 
Eg: A--->B-->c
  
3. Types of Graphs
1. Directed Graph ( also called Digraph)
* Edges have direction --> Like one -way Roads.
* A--> B--> C-->D
* there is no back direction keep on move to forward direction,
* Here you can go from A to B , But not from B to A 

2. Undirected Graph:
* edges have no directon --> like two way roads
* A<-->B<--->C<-->D
* You can go back and forth between A and B

3. Weighted Graph
* Each edges has a number ( called weight), Like distance or cost
* A--5--B--2--C
* Means Distance from A to B is 5
* From B to C is 2

4. Unweighted Graph
* Edges don't have any weight , Just connenctions

4. How to represent a Graph in code 
* There are 2 main Ways to represent in graph
1. Adjacency matrix (like a 2D Matrix)
* if you have N Nodes, Create an N x N Matrix.
* If there 's an edge from node i to node J, Put 1(Or the weight), Else ,0
* Matrix is symmertic for undirected Graph
* Good for: Checking if ab edge exists Quickly.

2. Adjacency List (Link a Dictionary)
* Store each node and a list os its neighbors
* Good for Saving space for Large Graph

5. Real- Life Use of Graphs
*  Google Map--> Cities as nodes , roads as edges
* Social Media --> Peoples as Nodes , Frienda/Follwers as Edges
* Internet--> Web pages as node, hyperlink as edges
* Recommendaion System-->items and user as nodes
* Airline Routes--> Airport as nodes , Fight are Edges.



```python
# let's implement a simple Graph class with basic functionalities
class Graph:
    def __init__(self):
        self.adjacency_list = {}

    def add_edge(self, u, v):
        if u not in self.adjacency_list:
            self.adjacency_list[u] = []
        if v not in self.adjacency_list:
            self.adjacency_list[v] = []
        self.adjacency_list[u].append(v)
        self.adjacency_list[v].append(u)  # For undirected graph

    def remove_edge(self, u, v):
        if u in self.adjacency_list and v in self.adjacency_list[u]:
            self.adjacency_list[u].remove(v)
        if v in self.adjacency_list and u in self.adjacency_list[v]:
            self.adjacency_list[v].remove(u)

    def get_neighbors(self, u):
        return self.adjacency_list.get(u, [])

    def __str__(self):
        return str(self.adjacency_list)
```

```python
# let's implement directed graph as well
class DirectedGraph:
    def __init__(self):
        self.adjacency_list = {}

    def add_edge(self, u, v):
        if u not in self.adjacency_list:
            self.adjacency_list[u] = []
        if v not in self.adjacency_list:
            self.adjacency_list[v] = []
        self.adjacency_list[u].append(v)

    def remove_edge(self, u, v):
        if u in self.adjacency_list and v in self.adjacency_list[u]:
            self.adjacency_list[u].remove(v)

    def get_neighbors(self, u):
        return self.adjacency_list.get(u, [])

    def __str__(self):
        return str(self.adjacency_list)
```

```python
# Lets implement undirected weighted graph
class WeightedGraph:
    def __init__(self):
        self.adjacency_list = {} 
    def add_edge(self, u, v, weight):
        if u not in self.adjacency_list:
            self.adjacency_list[u] = []
        if v not in self.adjacency_list:
            self.adjacency_list[v] = []
        self.adjacency_list[u].append((v, weight))
        self.adjacency_list[v].append((u, weight))  # For undirected graph
```
