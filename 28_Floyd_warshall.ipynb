# Floyd-warshall Algorithms
1. What is Floyd_Warshall
* Floyed-warshall is an algorithm used to find the shortest path between all paths of vertices in a weighted graph

* That means
* If you have n cities and some direct roads between them (with distance). this algorithm tells you the shortest path between every pair of cities, even if they areindiretly connected through other cities.

2. When to use Floyd-warshall
* Use floyd-warshall when:
* you want shortest distance between all pairs (not just one-to-one like Dijkstra)
* graph is smalll ( e.g <100 vertices ), Because it's not very fast for large graphs.
* you have negative weights, and Dijkstra doesn't work.

3. key terminologies
* Terms---> Meaning
* Graph ---> A set of points (vertices ) connected by line (edged)
* Vertex --> A node in the graph
* Edge--> A connection between two vertices
* weight --> Cost or Distance to move from one vertex to another
* Shortest path --> the minimum total weight to travel from one vertex to another
*  negative weight -->edge with a cost less than O
* Negative cycle --> A loop inthe graph where total cost become negative. Floyd -warshall cannot work correctly if this exists.

* Algorithms  intution (How it works in simple Terms)
* Let's say you want go to from city i to city J.
* You already know direct distance between somw cities.

* now you check:
* Is there any city k such that going I-->K-->J is shorter than directly going i-->j
* you do this check for every pair of cities and for every possiable middle city k.
* You repact this for all k from 0 to n-1.
* that's it simle idea, powerful result.

* After running Floyd-warshall we will get the shortest paths between all cities including indirect ones.


3. time and space complexity
* Matrics---> value---> Meaning
* Time comlexity -->O(N3)--> Because we use 3 nested loops for k,i, and j
* space complexity---> O(n2)--->Becuase we use a 2d List(matrix) to store distance.

4. Advantages 
* Easy to implement
* Works with negative weights
* gives shortest path between all pairs.
* Can detect Negative cycle ( by checking if graph[i][i]<0 after the algorithm)


5. Disadvantages
* Slow for large graph O(n^3) is too much for n > 100
* doesn't work with negative cycle ( result become invalid)

* final summary
* points --> Info
* Goal--> shortestt path between all pairs
* Works on --> Weighted graph (positive or negative)
* Fails on --> Graph with negative weight cycles
* Time --> O(n^3)
* space --> O(n^2)
* use when --> small graph, all pairs needed , megative edges allowed.

```python
# Simple code for floyd-warshall algorithm to find shortest paths in a weighted graph with positive or negative edge weights (but with no negative cycles)
# n= number of vertices
# graph= adjacency list representation of the graph where graph[u] = [(v1, weight
# ), (v2, weight2), ...] indicates an edge from u to v1 with weight weight1, and so on.
import sys
def floyd_warshall(graph):
    # Number of vertices in the graph
    num_vertices = len(graph)
    
    # Initialize distance matrix with graph values
    dist = [[float('inf')] * num_vertices for _ in range(num_vertices)]
    
    # Set distance from each vertex to itself to 0
    for i in range(num_vertices):
        dist[i][i] = 0
    
    # Set initial distances based on the input graph
    for u in range(num_vertices):
        for v, weight in graph[u]:
            dist[u][v] = weight
    
    # Floyd-Warshall algorithm
    for k in range(num_vertices):
        for i in range(num_vertices):
            for j in range(num_vertices):
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    
    return dist
```

```python
# Example usage
if __name__ == "__main__":
    # Example graph represented as an adjacency list
    graph = [
        [(1, 3), (2, 8), (4, -4)],# Edges from vertex 0
        [(3, 1), (4, 7)],# Edges from vertex 1
        [(1, 4)],#w Edges from vertex 2
        [(0, 2), (2, -5)],# Edges from vertex 3
        [(3, 6)]# Edges from vertex 4
    ]
    
    shortest_paths = floyd_warshall(graph)
    
    print("Shortest path matrix:")
    for row in shortest_paths:
        print(row)
```
