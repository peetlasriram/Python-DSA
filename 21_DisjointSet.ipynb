# Disjoint Set (Union Find)
1. What is a Disjoint Set (union-Find)
* Disjoint Set is a Data Structure used to keep track of elements grouped into sets.
* Where each element belongs to exactly one group(set)
* It is also called Union-Find Because it mainly supports two operations
* if Find(x) --> Which set does X belong to
* union(x,y)--> Join the sets containing X and y.

2. Real-life Example
* imagine you are forming friend groups
* Each person starts in their own group
* It two people become friends, you merge thier group
* Later you want to check whether two people are in the same groups.
* this is where Disjoint Set Helps.


3. When to use Disjoint set
* Use Disjoint set when
* you need to group elements dynamically
* you want to check if two elements are in the same group

* used in problems like:
* Cycle Detection in graph
* Kruskal's algorithm ( for minimum spanning tree)
* Connected components

4. Basic terminology and Functions
* 1. Find(x)
* Return the representative ( Leader) of the set that contains X.
* This helps check whether two elements are in the same set.
* 2. Union(x,y)
* joins the sets that conatins X and Y
* Naives implementation ( Basic Version Without Optimization)

3. optimizations
* When we do find(x), we also update x's parent to the root.
*this is make the tree flat and faster for future calls.

* Union By rank/size
* Instead of randomly joining sets.
* Attach the smaller tree under the bigger one to keep the structure short.

```python
# let's implement a Disjoinnt Set Union (DSU) data structure in Python
class DisjointSetUnion:
    def __init__(self, size):
        # Initialize the parent array and rank array
        self.parent = list(range(size))
        self.rank = [1] * size

    def find(self, u):
        # Path compression heuristic
        if self.parent[u] != u:
            self.parent[u] = self.find(self.parent[u])
        return self.parent[u]

    def union(self, u, v):
        # Union by rank heuristic
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

    def connected(self, u, v):
        # Check if two elements are in the same set
        return self.find(u) == self.find(v)
```

```python
# Example usage
if __name__ == "__main__":
    dsu = DisjointSetUnion(10)

    dsu.union(1, 2)
    dsu.union(2, 3)
    dsu.union(4, 5)

    print(dsu.connected(1, 3))  # Output: True
    print(dsu.connected(1, 4))  # Output: False

    dsu.union(3, 4)

    print(dsu.connected(1, 4))  # Output: True
```
