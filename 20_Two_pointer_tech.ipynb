# Two pointer Technique
1. What is twom points technique
* Two pointer is a technique where you are use two variable ( pointer) to move though an array. String.or list to solve problems faster and using less memory.
* insead of using two loops (which is slower). we use two pointers

2. The pointers can:
* Start at different positions
* Move in the same or opposite directions
*  move at different speeds
* The idea is o avoid extra space and reduce time.


3. When to use it.
* You should think about using two pointers when
* you need to process a list from both start and end
* you want to find pairs that match a condition
* you ness to reverse an array
* you need to remove dupilcate from a sorted list
* you want to find the middle of a list.
* you want to detect a cycle in a linked list
* you want to merge two sorted array
* you want to solve a problem without using extra space

4. Type of two pointers.
1. Left -Right Pointer
* How is it works
* One pointerr start at the left (start of list)
* one pointer start at the right( Start of list)
* Both move towars each other or based on condition

* Use Cases
* Check if a string is palindrome 
* Find two numbers in a sorted array that add up to target.
* Reverse an Array.
* Partition an array.




2. Fast-Slow pointer ( Tortoise  & Hare)
* How is it works:
* Both pointer start at the beginning.
* One moves slow (1 Step at a time)
* One moves fast ( 2 step at a time)

* Use Cases
* Find the middle of list
* Detect cycle in a linked list
* Remove duplicates in a sorted l  ist



4. Time and Space Complexity
* Type  | Time complexity | Space Complexity
* Left-Right pointer |  o(n)| o(1)
* Fast-Slow pointer | O(n)  | o(1)

* Advantages 
* Simple to implement
* Faster than using two loops(no o(n2))
* Saves space (o(N))
* Works really well with sorted array and linked list
* Help with tricky problems involving pairs or moving windows.

5. Common mistake to Avoid
* Not updating the pointer correctly --> infnity Loop
* Not handling edges cases (empty list, single element)
* using wrong conditions in while loop

* Worksss great for arrays, string, Linked list problems.

```python
# let's implement two pointers technique in python
# technique1 : Left and Right pointers to check for palindrome
def palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
```

```python
s="madam"
print(palindrome(s))  # Output: True
```

```python
# lett's implement two pointers technique in python
# technique2 : Fast and Slow pointers to find middle of linked list
def find_middle(s):
    slow, fast = 0, 0
    while fast < len(s) - 1:
        slow += 1
        fast += 2
    return s[slow]
```

```python
mid=[1,2,3,4,5]
print(find_middle(mid))  # Output: 3
```
