# Senerio:  Delivering Pizze
* You're a delivery persons and you need to delivery pizza to n house

*   O(1): Always Takes the time.
*   O(N): Grown linearly with input size.
*   O(n2): growns quadrantial with nested loops. 
*   O(log n): grown slowly cutting the problem in half each item.
*   O(2n) : Grown exponentially, doubling the work at every step.

# 1. Constant Time - O(1)
* If your boss ask you to delivery a pizza to the first house on the list.
* it does't matter how many house are on the city
* you go directly to that one house

```python
def constant_time(arry):
    return arry[0]
```

```python
arry=[1,2,3,4,5,6,]
print(constant_time(arry))
```

# 2.Logarthmic Time -O(Log n)
* If the house are organised in the tree-like structure (e.g, one main road branching intl smaller street)
* You can deliver faster by dividing your search

```python
def logarithmic_time(n):
    if n <= 1:
        return n
    else:
        return logarithmic_time(n // 2) + 1
```

```python
print(logarithmic_time(16))
```

# 3.Linear Time - O(n)
* If you need to deliver pizza to all the house, You will visit each house one by one.
* The time it takes increase with the number of houses.

```python
def linear_time(arry):
    for num in arry:
        print(num)
```

```python
print(linear_time(arry))
```

# 4.Quadratic Time O(n2)
* If you try to check every possiable pair of house to decide the shortest delivery route.
* The number of comparisons grows rapidly as the number of house increase.

```python
def quadratic_time(arry):
    for i in arry:
        for j in arry:
            print(i, j)
```

```python
print(quadratic_time(arry))
```

# 5.Exponsinal Time O(2N)

```python
def exponential_time(n):
    if n <= 1:
        return n
    else:
        return exponential_time(n - 1) + exponential_time(n - 1)
```

```python
print(exponential_time(5))
```
