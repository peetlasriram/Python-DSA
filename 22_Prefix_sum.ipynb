# Prefix_Sum
1. Waht is a Prefix Sum
*  Prefix sum is technique used to quickly calculate the sum of elements in range of an array or matrix.
* instead of calculating the sum from scratch every time , we precomputer the sum of elements.
* from the begining up each index.

2. This helps answer queries like:
* What is the sum index i to index j
* how many times has a number appered up to a point
* it Works for both 1D aarray (normal list) and 2D arrays (Grids/Matrices)

3. When to use Prefix Sum
* Without prefix sum, if you want to find the sum of a range (i...j).
* You need to loop from 1 to J every time -> Which is slow.

4. With prefix sum
* You compute a helper array once
* then answer any range sum in cinstant time (o(N))

5. 1D Prefix sum
* Goal: Create an array where prefix[i] stores the sum from index 0 to i
* Example :
* Arr=[2,4,6,8]
prefix sum=[2,6,12,20]

* Explantion
* Prefix[0]=2
* Prefix[1]=2+4=6
* Prefix[2]=2+4+6=12
* Prefix[3]=2+4+6+8=20

7. use cases of Prefix sum
* Range Sum quires
*  Frequent Count
* Count How many times a numner has appered up to a certain index.
* Subarrays with given sum
* Prefix sum helps tracks cumulative sums for indentifying subarrays that sum to a target.
* Checking Balance
* Find parts of an array that are equally weighted.

8. When to use Prefix sum.
* You need to answer multiple range sum queries.
* you want to optimize repeated sum calculations.

9. Time and Space complexity
* Term    | Meaning
* Prefix sum  : Running total of array/ matrix elements
* Range Query:  Asking sum between two indices or cells
* Cumulative Sum : Another Name for prefix sum 
* 2D Prefix sum : Prefix Sum for a matrix.

10. Adavantages
* Fast range sum Queries (o(1)time)
* Simple to implement with basic Logic
* Works well with large data when preprocessing is allowed.

11. Disadvantages
* Uses Extra space ( especial in 2D).
* Must recomputerr if the array/matrix changes.
* Not helpful when you need frequent updates to the original data.

```python
# lets create 2D prefix sum array in Python

def prefix_sum_2d(matrix):
    if not matrix or not matrix[0]:
        return []

    rows, cols = len(matrix), len(matrix[0])
    prefix = [[0] * (cols + 1) for _ in range(rows + 1)]

    for i in range(1, rows + 1):
        for j in range(1, cols + 1):
            prefix[i][j] = (matrix[i - 1][j - 1] +
                            prefix[i - 1][j] +
                            prefix[i][j - 1] -
                            prefix[i - 1][j - 1])
    return prefix
```

```python
# let's implement Example of prefix sum array in Python

def prefix_sum(arr):
    n = len(arr)
    prefix = [0] * (n + 1)
    for i in range(1, n + 1):
        prefix[i] = prefix[i - 1] + arr[i - 1]
    return prefix
```
