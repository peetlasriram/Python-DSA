# Dynamic Programming 
1. What is Dynamic Programming
* Dynamic programming (DP) is method to slove problems by breaking them into smallet parts, Solving each part only once, And reusing the answer later when  needed

2. Real-Life Analogy - Math Exam with Caching.
* Imagine you're taking a math exam, you solved a difficult multiplication in one questions.
* 34 X 41 == 1517
* Later, in another questions, You need the same calculation again.
* would you redo it? No
* you would just use the answer you already found -1517
* this is exactly what dp Does. It remember answer to smaller problems and reuse them to solve the big one faster.

3. why and when to use Dp
* Use Dp when the problem has:
1. Overlapping subproblems
* you solve the same small problem again and again.
* Example: In Fibonacci number, to find fib(5), you nees fib(4) and fib(3)-- but fib(3)--is used agin in fib(4)
2. Opitimal Substructure
* the final answer depens on opitimal answer to smaller problems
* Example :- To reach the minimum cost, you need to know the minimum cost of previous steps.
