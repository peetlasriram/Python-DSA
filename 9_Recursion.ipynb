# Recursion
1. What is recursion
* Recursion is a method in programming where a function calls itself to solve a problem. It breaks down a big problem into smaller subproblems untill it reaches the simplest case (base Case)

2. Defination 
* Recursion is a technique where a function calls itself untill it reaches a stopping condition(base case)

3. Real-Life Example : Box inside box approach
* Imgine you have big box, and inside that big box, there are smaller boxes, and inside each small box, there could't another box, and so on, Your task is to find a key inside the boxes

4. you open the big box
* if you find the key, you stop (base case)
* if you find another box  inside, you open that box and  repeat the process (Recusive case)
* you continue untill you find the way
* this is exactly how recursion works

5. Unstanding Recursion in python
* Basic Structure of Recursion

6. Every Recursion function has two parts
* Base Case :->The Condition that stops recursion
* Recursion Case :->The function calls itself with a smaller input.

# When to use Recursion
* When a problem can be divide into smaller subproblems of the same type
* when iterative solution hard to implement but recursion is natural
* when sloving problems related to trees, graphs, and backtracking

# When Not use for Recursion
* When stack memory is limited (recursion deapth can cause a stack overflow)
* when a loop-based solution is simpler and more efficient
* when performance matter (recursion might be slower than iteration)

```python
# Example of Recursion

def recursion(n):
    if n == 0 : #base case
        return 
    print(n)
    recursion(n - 1) #recursive case
```

```python
def factorial(n):
    print(f"calling factorial({n})")
    if n ==1:  # Base case
        print(f"returning 1 for factorial({n})")
        return 1
    reslut=n * factorial(n - 1)  # Recursive case
    print(f"returning {n} * factorial({n-1}) for factorial({reslut})") 
    return reslut
print(factorial(5))  # Output: 120
```

```python
print(factorial(5))  # Output: 120
```

```python
def fabinnaci(n):
    print(f"calling fabinnaci({n})")
    if n <= 0:  # Base case
        return 0
    elif n == 1:  # Base case
        return 1
    else:
        return fabinnaci(n - 1) + fabinnaci(n - 2)  # Recursive case
```

```python
print(fabinnaci(5))  # Output: 120
```
