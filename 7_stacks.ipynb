# Stacks
1. What is Stack
* A: Stack (Last in First Out)
* a stack works like a stack  of plates in your kitchen, The last plate you place on top is the first one you pick up-This is called LIFO (Last in First Out)

2. Real -life Example 
* Imagine a stack fo books, The book you place last on top is the first one you will pickup to read.

3. Main Operations in Stacks
* Push -> Add the element to the top.
* Pop -> Remove the top element,
* Peek/Top -> View the top element without removing it.
* IsEmpty -> Check if the stack is empty
4. Time complexity analysis for stack
* Push-> o(1) (adding at the end takes constant time)
* pop -> o(1)(Removing from the end takes constant time)
* Peek ->o(1)
*isEmpty->o(1)

```python
class  Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.items[-1]

    def size(self):
        return len(self.items)
```

```python
s=Stack()
s.push(1)   
s.push(2)
s.push(3)
print(s.pop())  # Output: 3
print(s.peek()) # Output: 2 
# A simple stack implementation in Python
print(s.size()) # Output: 2 
# A simple stack implementation in Python
# A simple stack implementation in Python
```
