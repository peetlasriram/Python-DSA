# Binary Search Tree
1. What is a binary Search Tree
* A Binary tree is a tree data structure where each node at most two childrens the topmost node is called the root,

2. Binary Search Tree(BST) : A special form of binary tree where:
* Every node's left subtree contains only nodes with values less than the root node value.
* Every node's right subtree contains only nodes with values greater than the root node value.
* Both the left and right subtree are also BSTs.

3. Each node contains
* Data( value)
* left (left child)
* Right (right child)

4. Common Terms
* Left node:A node with no childrens
* Hight: Number of edges from root node to deepest left.
* Full Binary Tree : Every node has 0 or 2 Childrens
* Perfect bInary tree : All internal node have 2 Childrens. And all leaves are at the same level.
* Level: A level start from 0 level every level has certain numbers.

6. Types of Traversals
* Inorder(LNR)
* Preorder(NLR)
* Postorder(LRN)

# Basic Operations
1. Search:
* If the value equals the root's value. return that root.
* if the value is less than root node, traverse the left subtree: if the greater than root node, traverse the right subtree.
* Time complexity: Average case O(log n), worst -case O(n)(in a skewed tree)

2. Insertion:
* similar to search: find the correct place where the node should reside, ensuring the ordering property is  insert the new node as a leaf.

3. Deletion:
* Three possible scenerios:
* node is a leaf : remove it
* node has one child: Replace the node with its child.
* node has two children:Find the inorder successor (or preorder). copy its value to the node. and then delete the successor (or predecessor). this keeps the BST properties intact.

4. Traversal:
* inorder traversal: visits  nodes in ascending order
* preorder and postorder traversal :useful for other  operations like copying a tree or evaluting expression trees.

```python
class node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
class binary_tree:
    def __init__(self):
        self.root = None
    def insert(self, data):
        if self.root is None:
            self.root = node(data)
        else:
            self._insert_rec(self.root, data)
    
    def _insert_rec(self, current_node, data):
        if data < current_node.data:
            if current_node.left is None:
                current_node.left = node(data)
            else:
                self._insert_rec(current_node.left, data)
        else:
            if current_node.right is None:
                current_node.right = node(data)
            else:
                self._insert_rec(current_node.right, data)
        
    def delete_tree(self):
        self.root = None 
    
    def diaplay(self):
        if self.root is not None:
            self._display_rec(self.root)
    def _display_rec(self, current_node):
        if current_node is not None:
            print(current_node.data, end=' ')
            self._display_rec(current_node.left)
            self._display_rec(current_node.right)

    # traversal methods inorder(self):

    def inorder(self):
        if self.root is not None:
            self._inorder_rec(self.root)
    def _inorder_rec(self, current_node):
        if current_node is not None:
            self._inorder_rec(current_node.left)
            print(current_node.data, end=' ')
            self._inorder_rec(current_node.right)

    def preorder(self):
        if self.root is not None:
            self._preorder_rec(self.root)

    def _preorder_rec(self, current_node):
        if current_node is not None:
            print(current_node.data, end=' ')
            self._preorder_rec(current_node.left)
            self._preorder_rec(current_node.right)
    
    def postorder(self):
        if self.root is not None:
            self._postorder_rec(self.root)

    def _postorder_rec(self, current_node):
        if current_node is not None:
            self._postorder_rec(current_node.left)
            self._postorder_rec(current_node.right)
            print(current_node.data, end=' ')
            
    def search(self, data):
        if self.root is not None:
            return self._search_rec(self.root, data)
        else:
            return False
    def _search_rec(self, current_node, data):
        if current_node is None:
            return False
        if current_node.data == data:
            return True
        elif data < current_node.data:
            return self._search_rec(current_node.left, data)
        else:
            return self._search_rec(current_node.right, data)
    
```
