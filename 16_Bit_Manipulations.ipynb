4. bitwise Operators
* These operators work bit by bit on number
* let's Say:
* A = 5 == 0101
* B = 3==  0011

1. & (And)
* A & B =0101-0011=0001->1
* only 1 if both bits are 1

2. | (OR)
* A | B == 0101-0011= 0111 -> 7
* if at least one of the btis is 1 

3. ^ (XOR)
* A ^ B == 0101- 00Q11 = 0110 -> 6
* 1 if bits are different 

4. ~ (NOT)
*  ~ A ==  ~0101= 1010 ->10 (in 4-bit world) - Gives negative number in signed int
* Filps all bits (0->1,1->0)

5. << (LEFT SHIFT)
* A <<1 == 0101 <<1010-10
* Multiplies by 2 

6. >> ( RIGHT SHIFT)
* A >> 1 ==0101>> 1==0010 -2
* Divides by 2

```python
# Bit manipulation tricks in Python
#.= Check if a number is even or odd
def is_even(n):
    if n & 1 == 0:
        return f"even is number :{n} "
    else:
        return "odd is number :" + str(n)
    return     
# Example usage
print(is_even(5))  # Output: True
```

```python
a=5
b=3

c=a & b  # Bitwise AND
d=a | b  # Bitwise OR
e=a ^ b  # Bitwise XOR
f=~a     # Bitwise NOT
g=a << 2 # Left shift
h=a >> 1 # Right shift
print(f"Bitwise AND: {c}")
print(f"Bitwise OR: {d}" )
print(f"Bitwise XOR: {e}")  
print(f"Bitwise NOT: {f}")
print(f"Left Shift: {g}")   
print(f"Right Shift: {h}")  
```

```python
binary_num = 26
print(f"Binary representation of {binary_num} is {bin(binary_num)}")
```
