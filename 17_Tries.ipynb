# Tries
1. What is Tries
* A Trie (pronounced "try") is a special tree-like Data structure used to store words or strings.
* Each node in the Trie Represent a character of word.
* It is also called a prefix tree because it store words based on their prefixes ( starting Letters).

2. Real-life Applications
1.  Autocomplete (like google search)
* When you type "app". it sugguest "apple","application",Etc. Tries help find all word start with "app"

2. Dictionary
* Store a large number of words in a way that makes it fast to check if a word exists or not.

3. Spell Checker
* Can be used to find misspelled words by checking if a word exists in the tries.

3. Trie Node Structure
* Each node in trie has:
* A mapor array to store child nodes(one for each letter a-z)
* A Boolean flag to mark the end of a word

4. hoe to works
* to insert a word like "cat", we gor from c->a->t->. and mark t as the end.
* To search "car", we follow c->a->r-> and check  if R is the end of the a word
* To Check if any word start with "do". we follow that path exists, the prefix is valid.

5. Time and space complexity

```python
# python code for trie data structure
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False
class Trie:
    def __init__(self):
        self.root = TrieNode()
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True
    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end_of_word
    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True
```

```python
# Example usage
trie = Trie()   
trie.insert("hello")
print(trie.search("hello"))  # True
print(trie.search("hell"))   # False
print(trie.starts_with("hell"))  # True

```
