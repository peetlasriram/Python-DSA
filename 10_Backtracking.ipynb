#   Backtracking
1.  What is Backtracking
* Backtracking is a way of trying all possible operations to solve a problems. But if one path does't work. we go back (undo) and try a difference path.
2. Think of it like exploring a maze
* you try one path. if you hit a dead end. you go back to the previous point and try a different path. you keep doing this untill you find the solution.

3.  how it works (simple steps)
*  we flow 3 basic steps
* Try: Make a choice 
* Recurise: Go Deeper with that choice (move forward)
* Backtrack: if it didn't work. undo the choice and try another.
* this is why it is called Back tracking

4. use cases(where is back tracking used)
*  Generating permutations and combination
* sloving Suduko
* Solving N-Queens Problems
* Solving maze or pazzle Problems
* Word Search on a board


5. visualize the decision tree
* let's say we want to find all possiable 2-letter words using A,B,C
* Step By Step
* Start-A=>AB-AC
* B=>BA->BC
* C=>CA->CB


```python
def backtracking(index,current,s):
    if index == len(s):
        print(current)
        return  
    backtracking(index + 1, current , s)
    backtracking(index + 1, current + s[index], s)
    
```

```python
s="ABC"
print(backtracking(0,"",s))
```
