```python
# Example All in one Code
Mod=10**9+7

def add(a,b):
    return (a+b)%Mod
def sub(a,b):
    return (a-b+Mod)%Mod
def mul(a,b):
    return (a*b)%Mod
def power(a,b):
    res=1
    while b:
        if b&1:
            res=mul(res,a)
        a=mul(a,a)
        b>>=1
    return res
def inv(a):
    return power(a,Mod-2)
def div(a,b):
    return mul(a,inv(b))

# Example Usage
a=7
b=5
print("Add:",add(a,b))
print("Sub:",sub(a,b))
print("Mul:",mul(a,b))
print("Div:",div(a,b))

# Modular Arithmetic Operations
print("Power:",power(a,b))

# Modular Inverse
print("Inverse of a:",inv(a))

# Modular Arithmetic Operations in Python
print("Add:",(a+b)%Mod)


print("Sub:",(a-b)%Mod)
print("Mul:",(a*b)%Mod)

print("Div:",(a*pow(b,Mod-2,Mod))%Mod)

# Modular Arithmetic Operations in Python

print("Power:",pow(a,b,Mod))

# Modular Inverse in Python
print("Inverse of a:",pow(a,Mod-2,Mod))


```

```python
# let implement extended Euclidean algorithm in python
def extended_gcd(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = extended_gcd(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y
# Example usage
a = 30
b = 12
gcd, x, y = extended_gcd(a, b)
print(f"GCD of {a} and {b} is {gcd}")
print(f"Coefficients x and y are {x} and {y} respectively")
```

```python
def mod_inverse(a, m):
    gcd, x, y = extended_gcd(a, m)
    if gcd != 1:
        raise ValueError("Inverse doesn't exist")
    else:
        return x % m
# Find modular inverse of a under modulo m
a = 3
m = 11
inverse = mod_inverse(a, m)
print(f"Modular inverse of {a} under modulo {m} is {inverse}")
# Extended Euclidean Algorithm to find GCD and coefficients
```
