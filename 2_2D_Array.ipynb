```python
def matrix1(matrix,m,n):
    for i in range(m):
        for j in range(n):
            print(matrix[i][j],end="|")
        print("")
```

```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
m=n=3
print("matrix")
matrix1(matrix,m,n)
```

# 1. Search in 2D matrix (leetcode Q74)

```python
def brutematri(matrix,t):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j]==t:
                return True
    return False
```

```python
def Brutematri(matrix,t):
    for rows in matrix:
        for cls in rows:
            if cls==t:
                return True
    return False
            
```

```python
def opimizedmatrix(matrix,t):
    if not matrix:
        return False
    row=len(matrix)
    col=len(matrix[0])
    left=0
    right=col-1
    while left<row and right>=0:
        if matrix[left][right]==t:
            return True
        elif matrix[left][right]>t:
            right-=1
        else:
            left+=1
    return False
```
