# 1. Search in 2D matrix (leetcode Q74)

```python
def bruteS2DM(matrix,t):
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if matrix[i][j]==t:
                return True
    return False
```

```python
def BruteS2DB(matrix,t):
    for rows in matrix:
        for cls in rows:
            if cls==t:
                return True
    return False
            
```

```python
def opimizedmatrix(matrix,t):
    if not matrix:
        return False
    row=len(matrix) # number of rows is 3
    col=len(matrix[0])# number of columns is 4
    left=0 # start from first row
    right=col-1 # start from last column
    
    while left<row and right>=0: # boundary check
        if matrix[left][right]==t: # found the target
            return True # found the target
        elif matrix[left][right]>t: # current element is greater than target
            right-=1 # move left
        else: # current element is less than target
            left+=1 # move down
    return False # target not found
```

```python
matrix=[[1,3,5,7],[10,11,16,20],[23,30,34,60]]
t=5
#output=True
print("Search  in matrix:",BruteS2DB(matrix,t))
print("Search  in matrix:",opimizedmatrix(matrix,t))
```

# 2. Transpose a Matrix (leet code Q867)

```python
def bruteTM(matrix):
    row=len(matrix)
    col=len(matrix[0])
    transposed=[[0]*row for _ in range(col)]
    for i in range(row):
        for j in range(col):
            transposed[j][i]=matrix[i][j]
    return transposed
```

```python
def optimizedTM(matrix):
    transposed=[[0 for _ in range(len(matrix))]for _ in range(len(matrix[0]))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            transposed[j][i]=matrix[i][j]
    return transposed
```

```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]   
# output=[1,4,7,2,5,8,3,6,9]
print("Transpose of matrix:",bruteTM(matrix))
print("Transpose of matrix:", optimizedTM(matrix))
```

# 3. Find Row with Maximum sum (leetcode )

```python
def bruteFRMS(matrix):
    max_sum=float('-inf')
    row_index=-1
    for i in range(len(matrix)):
        current_sum=sum(matrix[i])
        if current_sum>max_sum:
            max_sum=current_sum
            row_index=i+1
    print("Maximum sum is:",max_sum)
    print("Row index with maximum sum is:",row_index)
```

```python
def optimizedFRMS(matrix):
    max_row,max_sum= max(enumerate(map(sum,matrix)),key=lambda x:x[1])
    print("Maximum sum is:",max_sum)
    print("Row index with maximum sum is:",max_row+1)
    print(list(enumerate(map(sum,matrix))))
```

```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]   
   
# output=[(0,6),(1,15),(2,24)]
#print("find row with max 1s:",bruteFRMS(matrix))
print("find row with max 1s:",optimizedFRMS(matrix))
```

# 4. snake in matrix (leetcode Q3248)

```python
def bruteSM(n,commands):
    matrix=[[0 for _ in range(n)]for _ in range(n)]
    x=y=0
    matrix[x][y]=1
    for command in commands:
        if command=="RIGHT":
            y+=1
        elif command=="LEFT":
            y-=1
        elif command=="UP":
            x-=1
        elif command=="DOWN":
            x+=1
        matrix[x][y]=1
    return matrix
```

```python
def opitimizedSM(n,commands):
    row,col=0,0
    for command in commands:
        if command=="RIGHT":
            col+=1
        elif command=="LEFT":
            col-=1
        elif command=="UP":
            row-=1
        elif command=="DOWN":
            row+=1
    return (row*n)+col
```

# 5. Set martix Zeros ()

```python
def bruteSZ(matrix):
    m,n=len(matrix),len(matrix[0])
    rows=set()
    cols=set()  
    for i in range(m):
        for j in range(n):
            if matrix[i][j]==0:
                rows.add(i)
                cols.add(j)
    for r in rows:
       for i in range(n):
           matrix[r][i]=0
    for c in cols:
        for j in range(m):
            matrix[j][c]=0
    return matrix
```

# 6. Max Subarray and kadene's Algorithms


```python
def bruteMSS(matrix):
    max_sum=float('-inf')
    current_sum=0
    for num in matrix:
        current_sum+=num
        if current_sum>max_sum:
            max_sum=current_sum
        if current_sum<0:
            current_sum=0
    return max_sum
```

```python
matrix=[-2,1,-3,4,-1,2,1,-5,4]

print("maximum subarray sum:",bruteMSS(matrix))
```

# 7. Spiral matrix | Leetcode Q54

```python
def bruteSM(matrix):
    result=[] # to store the spiral order
    if not matrix: # empty matrix
        return result # return empty list
    top,bottom,left,right=0,len(matrix)-1,0,len(matrix[0])-1 # boundaries
    while top<=bottom and left<=right: # while boundaries are valid
        for i in range(left,right+1): # traverse from left to right
            result.append(matrix[top][i]) # add to result
        top+=1 # move down
        for i in range(top,bottom+1): # traverse from top to bottom
            result.append(matrix[i][right]) # add to result
        right-=1 # move left
        if top<=bottom: # check if there are rows remaining
            for i in range(right,left-1,-1): # traverse from right to left
                result.append(matrix[bottom][i]) # add to result
            bottom-=1 # move up
        if left<=right: # check if there are columns remaining
            for i in range(bottom,top-1,-1): # traverse from bottom to top
                result.append(matrix[i][left]) # add to result
            left+=1 # move right
    return result # return the spiral order
```

```python
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
#output=[1,2,3,6,9,8,7,4,5]
print("spiral matrix traversal:",bruteSM(matrix))
#print("spiral matrix traversal:",optimizedSM(matrix))
```
