# Queue
1. what is queue
* A Queue is data structure that follws the first in first out (FIFO) priciple, this means the first element added to the queue will be the first one to be removed - just like people standing in a line for tickets.
2. Real-Life Example
* imgine you are wating in a line at a movie theater:
* The first person who joined the line will be the first person to get the ticket and leave.
* The person who joined the line will have to wait untill every one before them gets their tickets.
* this "line " behavior is exactly how queue works in program
3. basic Queue operations
* Enqueue(add): Add an element to the back of the queue
* Dequeue (Remove): Remove an element from the front of the queue
* Front(Peek): View the element at the front without removing it.
* isEmpty: Check if the queue is empty.

1. Enqueue
2. Dequeue
3. Front
4. isEmpty

```python
class Qeueue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self.items.pop(0)
    def front(self):
        if self.is_empty():
            raise IndexError("front from empty queue")
        return self.items[0]

    def size(self):
        return len(self.items)
```

```python
a=Qeueue()
a.enqueue(1)        
a.enqueue(2)
a.enqueue(3)
print(a.dequeue())  # Output: 1
```
