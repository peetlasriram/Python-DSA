# HEAP
1. What is heap
*   A heap is special type of binary tree (a tree where each node has at most two children).
* I t is mostly used to implement priority queues.

2. Real-Life Example:
* Think of a hospital emergency room:
* Patients are added (heappush) as they arrived.
* The one with highest priority (most serious condition) is treated first (heappop.)

3. Type of Heaps
* -> Min-Heap
* The smallest element is at the root (top)
* All parents nodes are smaller than their children
* Eg : root node 2 and child node 5 and 7

* -> Max-Heap
* the largest element is at the root Node.
* All parenets node are greater than theit children nodes.
* Eg:- parent node 9 and child nodes are 2 and 4.

```python
import heapq
# python's has heapq module implements a min-heap by default.
# To simulate a max-heap, we can invert the values by multiplying them by -1

# let's create a max-heap using heapq
class MaxHeap:
    def __init__(self):
        self.heap = []

    def push(self, item):
        # Invert the value to simulate max-heap
        heapq.heappush(self.heap, -item)

    def pop(self):
        # Invert the value back when popping
        return -heapq.heappop(self.heap)

    def peek(self):
        # Peek at the maximum value without removing it
        return -self.heap[0] if self.heap else None

    def is_empty(self):
        return len(self.heap) == 0

    def size(self):
        return len(self.heap)
```

```python
# Example usage
if __name__ == "__main__":
    max_heap = MaxHeap()
    max_heap.push(10)
    max_heap.push(20)
    max_heap.push(5)

    print("Max element:", max_heap.peek())  # Output: Max element: 20

    while not max_heap.is_empty():
        print("Popped element:", max_heap.pop())
```

```python
# let"s Dijkstra's algorithm be implemented using this MaxHeap
def dijkstra(graph, start):
    max_heap = MaxHeap()
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    max_heap.push(0)

    while not max_heap.is_empty():
        current_distance = -max_heap.pop()  # Invert back to get the actual distance

        for neighbor, weight in graph[current_distance].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                max_heap.push(-distance)  # Invert to maintain max-heap property

    return distances
```

```python
# Example usage
if __name__ == "__main__":
    graph = {
        0: {1: 4, 2: 1},
        1: {3: 1},
        2: {1: 2, 3: 5},
        3: {}
    }
    start_node = 0
    distances = dijkstra(graph, start_node)
    print("Distances from start node:", distances)
```

```python
# lett's OS task scheduling simulation using MaxHeap
class TaskScheduler:
    def __init__(self):
        self.task_heap = MaxHeap()

    def add_task(self, task_name, priority):
        # Push task with its priority (inverted for max-heap)
        self.task_heap.push((priority, task_name))

    def get_next_task(self):
        if not self.task_heap.is_empty():
            priority, task_name = self.task_heap.pop()
            return task_name
        return None # No tasks available
```
