# Topological Sort
1. What is Topological Sort
* Topological Sort is a linear ordering of the vertices of a Directed cycle Graph (DAG) such that for every directed edge U-->V, Node U comes before node V in the Ordering.

2. When is Topological Sorting Applicable.
* Only work on Directed Acycle Graph (DAG)
* If the graph has a cycle -> Topological sort is not Posssiable.

3. Real-Life Example /Use Cases
* Usage:--> Explanation
* Task Scheduling :-->Some tasks must be done before Other
* Course Prerequisites:--> Take Course A before Course B
* Build Systems:--> Compile Dependencies in the correct order
* Job scheduling :--> Jobs With dependency Constraints.

4. Two Approaches to Topological Sort
* We can do topological sorting using either
* DFS(Depth First Search)
* kahn's Algorithm(BFS)

* 1. DFS-based Topological Sort
* * Perform DFS on Every unvisited node
* * Once DFS is finished for a node, add it to the Result list.
* * Finally.Reverse the result list.

* 2. Kah's Algorithms(BFS-Based)
* * Count in-degrees of all nodes (Number of incoming edges)
* * Start With nodes that have in-degree 0 (no Dependencies)
* * use a queue to preocess nodes and reduce in-degree of neighbors.
* * if neighbor's in-degree 0--> add it to queue
* * Collect Nodes in order

```python
#DFS-Based Topological Sort Implementation in Python4
# DFS-based Topological Sort
# Perform DFS on Every unvisited node
# Once DFS is finished for a node, add it to the Result list.
#Finally.Reverse the result list.
def topoSortDFS(graph,V):
    visited = [False] * V
    stack = []

    def dfs(v):
        visited[v] = True
        for neighbor in graph[v]:
            if not visited[neighbor]:
                dfs(neighbor)
        stack.append(v)

    for i in range(V):
        if not visited[i]:
            dfs(i)

    stack.reverse()# Reversing the stack to get the correct topological order
    return stack
```

```python
#lets implement Topological Sort using Kahn's Algorithm
from collections import defaultdict, deque
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.indegree = defaultdict(int)

    def add_edge(self, u, v):
        self.graph[u].append(v)
        self.indegree[v] += 1
        if u not in self.indegree:
            self.indegree[u] = 0

    def topological_sort(self):
        queue = deque()
        for node in self.indegree:
            if self.indegree[node] == 0:
                queue.append(node)

        topo_order = []
        while queue:
            current = queue.popleft()
            topo_order.append(current)

            for neighbor in self.graph[current]:
                self.indegree[neighbor] -= 1
                if self.indegree[neighbor] == 0:
                    queue.append(neighbor)

        if len(topo_order) != len(self.indegree):
            raise ValueError("Graph has at least one cycle, topological sort not possible")

        return topo_order
```

```python
def topological_sort_example():
    g = Graph()
    g.add_edge(5, 2)
    g.add_edge(5, 0)
    g.add_edge(4, 0)
    g.add_edge(4, 1)
    g.add_edge(2, 3)
    g.add_edge(3, 1)

    try:
        order = g.topological_sort()
        print("Topological Sort Order:", order)
    except ValueError as e:
        print(e)
```
