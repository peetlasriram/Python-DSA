# Math
1. Prime Number
* A Prime number is a number greater than 1 that can only be divided by 1 and itself.
* Example: 2,3,5,7,11,13..
* Think of number as a special treasure box. Prime number only have two keys to open it, The key "1" and  the key thats's number itself. if there are other keys. it's not prime checking if number is prime: You can dividing it by number from 2 upto its square root.
* if none divide it exactly =,it 's prime
*  Time complexity: About O(root sqeare N)- Because you only check upto its sqareroot.
* Space complexity : O(N) only a few variables to keep track.
* Why improtant 
* used in cryptography, hashing and many algorithms.

2. GCD ( Greatest Common Divisor )
* The gcd is the largest number that divides two or more number exactly (without leaving a remainder )
* Example: GCD of 12 and 18 is 6.
* How to find GCD: use the euclidean algorithm-- Keep subtracking or tracking remainders untill zero.
* Time complexity: around o (log min(a,b))-- very fast.
* space complexity: O(1)
* Why important
* used to simpilify fractions, solve ratio problems, and in algorithms.

3. LCD (Least common multiple)
* The LCM is the smallest number that is diviable by 2 or more numbers
* Example: LCM OF 4 AND 6 IS 12
* personal takes 4 minutes to cover 6km and person2 takes 6 minutes to cover the same.
* they exactly meet after 12 minutes.
* Formula: LCM(a,b)= (aXb)/GCD(a,b)
* Time complexity : Same as GCD --O(log min(a,b))
* space complexity:O(1)
* Why is impotant
* Helps with scheduling. timing problems and combing cycles.

* Relations:
* LCM(a,b)=(axb)/GCD(a,b)

3. Sieve of Eratosthenes
* * A fast method to find all prime numbers upto a given numbers.
* how it works
1. Start with a list of numbers from 2 to n.
2. Mark multiple of each prime as not prime.
3. the number that are not marked are primes.
* time complexity: About (n log log n)--very efficient for big n
* space complexity : O(n) -- You need a list to mark numbers as crossed or not
* Why important
* helps quickly find prime, especially for large numbers.

5. Combinatorics (nCr and Factorials)
* Factorial (n!): the product of all number from 1 to n.
* Example: 4! = 4x3x2x1=24
* Time complexity :o(n) -- multiply number from 1 to n.
* space complexity :o(1)

ncr(Combination): Number of ways to chhosw r time from n items.
* formula=nCr=n!/r!x(n-r)!
* time complexity: depens on factorial calculations usually o(n)

* Why important
